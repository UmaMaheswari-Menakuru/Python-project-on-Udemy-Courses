## Fetching Required Columns and dropping unwanted 

```python
df1=df.drop(columns=['course_id','course_title','url','content_duration','published_timestamp'])
df1
```

## Feature Engineering

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
le=LabelEncoder()
```

### Converting categoriacl(ordinal data) to numerical

```python
df1['level']=le.fit_transform(df['level'])
df1['subject']=le.fit_transform(df['subject'])
```

```python
df1['is_paid']=df1['is_paid'].astype(int)
```

```python
df1
```

## Visualizations

```python
df['price'].value_counts()
```

### Histogram ----> price values

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=10, kde=True)
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### plotting maximum no.of subscribers for each upto level

```python
num_sub=df.groupby('level')['num_subscribers'].max()
```

```python
num_sub
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=df['level'], y=df['num_subscribers'], data=num_sub)
plt.title('Maximum Number of Subscribers by Level')
plt.xlabel('Level')
plt.ylabel('Max Number of Subscribers')
```

### plotting total no.of lectures given for each subject

```python
num_lec=df.groupby('subject')['num_lectures'].sum()
```

```python
num_lec
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=df['subject'], y=df['num_lectures'], data=num_sub)
plt.title('Maximum Number of lectures by Level')
plt.xlabel('Subject')
plt.ylabel('Lectures')
```

### plotting no.of subjects included in udemy course data set

```python
df['subject'].value_counts()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x=df['subject'],data=df1)
plt.title('Subjects for each course')
plt.xlabel('Subjects')
plt.ylabel('Courses')
plt.show()
```

 ### plotting no.of reviews given by subscribers for each level (positive and negative reviews)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='level', y='num_reviews', data=df)
plt.title('Number of Reviews by Level')
plt.xlabel('Level')
plt.ylabel('Number of Reviews')
plt.show()
```

### plotting pie chart to know percent of how many are paid and unpaid

```python
df['is_paid'].value_counts()
```

```python
plt.pie(x=df['is_paid'].value_counts(),labels=['True','False'])
plt.show()
```

```python
plt.pie(x=df['is_paid'].value_counts(),labels=['True','False'],autopct='%2.2f%%',explode=[0,0.3],shadow=True)
plt.show()
```

### Correlation

```python
corr=df1[['is_paid','price','num_subscribers','num_reviews','num_lectures','level','subject']].corr()
corr
```

```python
sns.heatmap(corr,annot=True)
plt.show()
```

### Covariance

```python
cov=df1[['is_paid','price','num_subscribers','num_reviews','num_lectures','level','subject']].cov()
```

```python
cov
```

```python
sns.pairplot(data=cov)
plt.show()
```

### Plotting no.of subscribers VS price based on is_paid category

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='price', y='num_subscribers', data=df, hue='is_paid')
plt.title('Number of Subscribers vs Price', fontsize=16)
plt.xlabel('Price')
plt.ylabel('Number of Subscribers')
plt.show()
```
